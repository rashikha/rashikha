```python
def q25(x):
  return x.quantile(0.25)
def q75(x):
  return x.quantile(0.75)
def iqr(x):
  return x.quantile(0.75) - x.quantile(0.25)
def lcl(x):
  return x.quantile(0.25) - (1.5 * (x.quantile(0.75) - x.quantile(0.25)))
def ucl(x):
  return x.quantile(0.75) + (1.5 * (x.quantile(0.75) - x.quantile(0.25)))

```

```python
quant_values_df = aerofit_data.groupby('Product').agg({'Age' : [q25,q75,iqr,lcl,ucl],'Income':[q25,q75,iqr,lcl,ucl], 'Miles' : [q25,q75,iqr,lcl,ucl]})
quant_values_df.columns = ['_'.join(col) for col in quant_values_df.columns]
quant_values_df.reset_index(inplace = True)
```

```python
quant_values_df
```

```python
quant_values_df.index = quant_values_df['Product']
```

```python
#aerofit_data.loc[(aerofit_data['Product'] == p) & ((aerofit_data['Age'] < quant_values_df.loc[p,'Age_lcl']) | (aerofit_data['Age'] > quant_values_df.loc[p,'Age_ucl']))]
```

```python
product_ls = aerofit_data['Product'].unique()
#outlier_count = {}
outliers_age = pd.DataFrame()

for p in product_ls:
   out_Age = pd.DataFrame(aerofit_data.loc[(aerofit_data['Product'] == p) & ((aerofit_data['Age'] < quant_values_df.loc[p,'Age_lcl']) | (aerofit_data['Age'] > quant_values_df.loc[p,'Age_ucl'])),aerofit_data.columns])
   out_Income = pd.DataFrame(aerofit_data.loc[(aerofit_data['Product'] == p) & ((aerofit_data['Income'] < quant_values_df.loc[p,'Income_lcl']) | (aerofit_data['Income'] > quant_values_df.loc[p,'Income_ucl'])),aerofit_data.columns])
   out_Miles = pd.DataFrame(aerofit_data.loc[(aerofit_data['Product'] == p) & ((aerofit_data['Miles'] < quant_values_df.loc[p,'Miles_lcl']) | (aerofit_data['Miles'] > quant_values_df.loc[p,'Miles_ucl'])),aerofit_data.columns])
   
   #outlier_count[p] = [len(out_Age),len(out_Income),len(out_Miles)]
   outliers_age.loc[p,'Age_outlier_count'] = len(out_Age)
   outliers_age.loc[p,'Income_outlier_count'] = len(out_Income)
   outliers_age.loc[p,'Miles_outlier_count'] = len(out_Miles)

outliers_age.reset_index(inplace=True)
outliers_age.rename({'index' : 'Product'},axis = 1)
```

```python
#What is the probability of a male customer buying a KP781 treadmill?
```

```python
aerofit_data.head()
```

```python
product_count_genderBased = pd.DataFrame(aerofit_data.groupby(['Gender','Product'])['Age'].count())
product_count_genderBased.reset_index(inplace = True)
product_count_genderBased.rename({'Age' : 'Product count'},axis = 1,inplace = True)

total_female_cust = product_count_genderBased.loc[product_count_genderBased['Gender'] == 'Female','Product count'].sum()
total_male_cust = product_count_genderBased.loc[product_count_genderBased['Gender'] == 'Male','Product count'].sum()

product_count_genderBased['prob_buyProduct_given_gender'] = product_count_genderBased.apply(lambda x,total_female_cust,total_male_cust : (x['Product count']/total_female_cust) if x['Gender'] == 'Female' else (x['Product count']/total_male_cust),axis = 1,args = [total_female_cust,total_male_cust])
```
